```python
import pandas as pd
```

```python
asylum_seeker = pd.read_csv('asylum_seekers.csv')
```

```python
asylum_seeker.replace('*', 'NaN', inplace=True)
asylum_seeker
```

```python
asylum_seeker.set_index('Year', inplace=True)
asylum_seeker
```

```python
asylum_seeker = asylum_seeker[['Origin', 'Applied during year', 'Total decisions']]
asylum_seeker.head()
```

```python
asylum_seeker_dtypes = {
    'Origin': 'str',
    'Applied during year': 'float',
    'Total decisions': 'float'
}

asylum_seeker= asylum_seeker.astype(asylum_seeker_dtypes)
```

```python
asylum_seeker_groupby = asylum_seeker.groupby(['Year', 'Origin']).sum()
asylum_seeker_groupby
```
