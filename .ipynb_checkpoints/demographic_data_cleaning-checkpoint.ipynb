```python
demographic_no_asterisks = demographic_df.replace("*", "NaN")
demographic_no_asterisks["Male 12-17"].value_counts()
```

```python
demo_dtypes_df = demographic_no_asterisks.astype(dtype = {"Year": str,
            "Country / territory of asylum/residence": str,
                                   "Female 0-4": float,
                                   "Female 5-11": float,
                                   "Female 5-17": float,
                                   "Female 12-17": float,
                                   "Female 18-59": float,
                                   "Female 60+": float,
                                   "F: Unknown": float,
                                   "F: Total": float,
                                   "Male 0-4": float,
                                   "Male 5-11": float,
                                   "Male 5-17": float,
                                   "Male 12-17": float,
                                   "Male 18-59": float,
                                   "Male 60+": float,
                                   "M: Unknown": float,
                                   "M: Total": float,
                                  } )
```

```python
demo_dtypes_df.dtypes
```

```python
demo_dtypes_df.head()
```

```python
len(demo_dtypes_df.index)
```

```python
demo_dtypes_no_nans = demo_dtypes_df.dropna(axis=0, how="any")
# len(demo_dtypes_no_nans.index)
demo_dtypes_no_nans.head()
```

```python
demo_by_year_residence = demo_dtypes_df.groupby(by=["Year", "Country / territory of asylum/residence"]).sum()
demo_by_year_residence.head(50)
```

```python
type(demo_by_year_residence)
```

```python
demo_by_year_residence.reset_index(inplace=True)
demo_by_year_residence.head()
```

```python
print(max(demo_by_year_residence["Year"]))
```

```python
print(min(demo_by_year_residence["Year"]))
```

```python
compact_demographics_df = demo_by_year_residence[["Year", 
                                                  "Country / territory of asylum/residence", 
                                                  "F: Total", 
                                                  "M: Total"]]
compact_demographics_df.head()
```
